```python
# Imports
import re
import geonamescache
import pandas as pd
```

```python
# Open the headlines file and insert them into a list
file = open("data/headlines.txt", 'r')
headlines_list = []
for line in file.readlines():
    headlines_list.append(line.strip())
file.close()
```

```python
# Creates a GeonamesCache object
gc = geonamescache.GeonamesCache()

# List of country names
countries_dict = gc.get_countries_by_names()
countries = countries_dict.keys()
```

```python
# List of cities names
cities_dict = gc.get_cities()
cities = []
for city in cities_dict: 
    if len(cities_dict[city]['name']) > 2: # Avoid 'Of' names
        cities.append(cities_dict[city]['name'])
```

```python
# Create headlines dataframe
headlines = pd.DataFrame(headlines_list, columns=['headline'])
headlines.head()
```

```python
# Country name regex
country_name_regex = r"("+r"\b|\b".join(countries)+r")"
```

```python
# City name regex
city_name_regex = r"("+r"\b|\b".join(cities)+r")"
```

```python
for index, row in headlines.iterrows():
    # Matching any country names
    matched_countries = re.search(country_name_regex, row.headline, flags=re.IGNORECASE)
    if matched_countries is not None:
        headlines.at[index, 'countries'] = matched_countries[0]
        
    # Matching any City names
    matched_cities = re.search(city_name_regex, row.headline, flags=re.IGNORECASE)
    if matched_cities is not None:
        headlines.at[index, 'cities'] = matched_cities[0]        
```
